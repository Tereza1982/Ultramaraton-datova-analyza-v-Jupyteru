```python
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('TWO_CENTURIES_OF_UM_RACES.csv').dropna(how='all', axis=1)
```

```python
df.head(10)
```

```python
df.shape
```

```python

CZ_nad_50km = df[
    ((df['Event distance/length'] > '50km') & (df['Event distance/length'] > '50mi')) &
    (df['Athlete country'] == 'CZE')
]

CZ_nad_50km.head(7)
```

```python
CZ_nad_50km['Event distance/length'].unique()
```

```python
CZ_nad_50km.info()
```

```python
CZ_nad_50km['Event name'] = CZ_nad_50km['Event name'].apply(str)
CZ_nad_50km
```

```python
CZ_nad_50km['Event name'] = CZ_nad_50km['Event name'].fillna('').str.split('(').str.get(1).str.split(')').str.get(0)
CZ_nad_50km
```

```python
CZ_nad_50km["Athlete age category"].unique()
```

```python
CZ_nad_50km[CZ_nad_50km["Athlete age category"] == 'M95'] # odhaleni chyby v udajich - bylo mu 88 let, ne 95
```

```python
# prevedeni data narozeni z float na integer a vytvoreni sloupce s aktualnim vekem ucastnika
CZ_nad_50km = CZ_nad_50km.dropna(subset=['Athlete year of birth'])
CZ_nad_50km['Athlete year of birth'] = CZ_nad_50km['Athlete year of birth'].astype('int')
CZ_nad_50km['Age'] = datetime.now().year - CZ_nad_50km['Athlete year of birth']
CZ_nad_50km # odhaleni dalsich chyb v dataframu - kategorie neodpovidaji veku
```

```python
# odstraneni sloupcu: Athlete country, Athlete ID
CZ_nad_50km.drop( ['Athlete country', 'Athlete ID'], axis=1)
```

```python
# Sice je sloupec dropnuty, ale chci na zaklade veku urcit vekovou kategorii u Nan hodnot
df['Athlete age category'].unique()
```

```python
def calculate_age_category(age):
    if age < 30:
        return '20'
    elif age < 40:
        return '30'
    elif age < 50:
        return '40'
    elif age < 60:
        return '50'
    else:
        return '70+'

CZ_nad_50km['Athlete age category'] = CZ_nad_50km['Athlete age category'].fillna(CZ_nad_50km['Age'].apply(calculate_age_category))
CZ_nad_50km
```

```python
# cisteni Nan hodnot
CZ_nad_50km.isna().sum()
```

```python
CZ_nad_50km['Athlete club'] = CZ_nad_50km['Athlete club'].fillna('Unknown')
CZ_nad_50km.isna().sum()  # nezbyly zadne Nan hodnoty
```

```python
# kontrola duplicit
CZ_nad_50km[CZ_nad_50km.duplicated() == True]
```

```python
# resetovani indexu
CZ_nad_50km = CZ_nad_50km.reset_index(drop = True)
```

```python
CZ_nad_50km['Athlete average speed'] = CZ_nad_50km['Athlete average speed'].apply(float)
```

```python
# prejmenovani sloupcu
CZ_nad_50km = CZ_nad_50km.rename(columns = {
'Year of event':'year',
'Event dates':'race_day',                
'Event name': 'race_name',                   
'Event distance/length': 'race_length',         
'Event number of finishers': 'race_number_of_finishers',      
'Athlete performance':'athlete_performance',           
'Athlete club':'club',                  
'Athlete country': 'athlete_country',               
'Athlete year of birth': 'year_of_birth',          
'Athlete gender': 'gender',                
'Athlete age category': 'age_category',         
'Athlete average speed': 'average_speed',                          
'Age': 'age'})
```
